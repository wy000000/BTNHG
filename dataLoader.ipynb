```python
heteroData=dataset._heteroData
# 验证标签分布是否相同
print("\n=== 训练集和测试集标签分布验证 ===")
train_labels = heteroData['address'].y[heteroData['address'].train_mask]
test_labels = heteroData['address'].y[heteroData['address'].test_mask]

# 计算训练集标签分布
train_label_counts = torch.bincount(train_labels[train_labels != -1])
train_label_dist = train_label_counts.float() / train_label_counts.sum()

# 计算测试集标签分布
test_label_counts = torch.bincount(test_labels[test_labels != -1])
test_label_dist = test_label_counts.float() / test_label_counts.sum()

print(f"训练集大小: {len(train_labels[train_labels != -1])} ({len(train_labels[train_labels != -1])/heteroData.num_labeled:.2%})")
print(f"测试集大小: {len(test_labels[test_labels != -1])} ({len(test_labels[test_labels != -1])/heteroData.num_labeled:.2%})")
print("\n训练集标签分布:")
for i, count in enumerate(train_label_counts):
	if count > 0:
		print(f"标签 {i}: {count} 个 ({train_label_dist[i]:.2%})")

print("\n测试集标签分布:")
for i, count in enumerate(test_label_counts):
	if count > 0:
		print(f"标签 {i}: {count} 个 ({test_label_dist[i]:.2%})")
```
